```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error

# Load the dataset
file_path = r"C:\Users\Mr Maziya\Downloads\institutions.csv"  # Adjust path as needed
data = pd.read_csv(file_path)

# Assuming 'date' is the date column and 'savings' is the column for long-term savings
data['date'] = pd.to_datetime(data['date'])  # Ensure date is in datetime format
data.set_index('date', inplace=True)

# Resample the data to monthly frequency, summing savings
monthly_data = data['savings'].resample('M').sum()

# Check for stationarity
def test_stationarity(timeseries):
    result = adfuller(timeseries)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    
test_stationarity(monthly_data)

# Fit the ARIMA model (you might need to adjust p, d, q values)
model = ARIMA(monthly_data, order=(1, 1, 1))  # Adjust orders as necessary
model_fit = model.fit()

# Print model summary
print(model_fit.summary())

# Make forecast
forecast = model_fit.forecast(steps=12)  # Forecast the next 12 months
forecast_index = pd.date_range(start=monthly_data.index[-1] + pd.DateOffset(1), periods=12, freq='M')

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(monthly_data, label='Historical Savings', color='blue')
plt.plot(forecast_index, forecast, label='Forecasted Savings', color='orange')
plt.title('Forecast of Long-Term Savings Growth')
plt.xlabel('Date')
plt.ylabel('Savings Amount')
plt.legend()
plt.show()

# Evaluate the model (if you have a test set)
# actual = ...  # Actual values for the forecasted period
# mse = mean_squared_error(actual, forecast)
# print(f'Mean Squared Error: {mse}')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
file_path = r"C:\Users\Mr Maziya\Downloads\institutions.csv"  # Adjust path as needed
data = pd.read_csv(file_path)

# Replace 'ActualDateColumnName' and 'ActualSavingsColumnName' with the actual column names
data['ACTIVE'] = pd.to_datetime(data['ACTIVE'])
data.set_index('ACTIVE', inplace=True)

# Resample the data to monthly frequency, summing savings
monthly_data = data['ACTIVE'].resample('M').sum()

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(monthly_data.values.reshape(-1, 1))

# Prepare the data for LSTM
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data) - time_step - 1):
        a = data[i:(i + time_step), 0]
        X.append(a)
        Y.append(data[i + time_step, 0])
    return np.array(X), np.array(Y)

# Reshape the data into a format for LSTM
time_step = 3
X, y = create_dataset(scaled_data, time_step)
X = X.reshape(X.shape[0], X.shape[1], 1)  # Reshape to [samples, time steps, features]

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Fit the model
model.fit(X, y, epochs=100, batch_size=32)

# Make predictions
predictions = model.predict(X)
predictions = scaler.inverse_transform(predictions)

# Prepare future predictions
future_steps = 12
future = []
last_data = scaled_data[-time_step:].reshape(1, time_step, 1)

for _ in range(future_steps):
    pred = model.predict(last_data)
    future.append(pred[0][0])
    last_data = np.append(last_data[:, 1:, :], [[pred]], axis=1)

# Inverse transform future predictions
future = scaler.inverse_transform(np.array(future).reshape(-1, 1))

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(monthly_data.index, monthly_data, label='Historical Savings', color='blue')
future_index = pd.date_range(start=monthly_data.index[-1] + pd.DateOffset(1), periods=future_steps, freq='M')
plt.plot(future_index, future, label='LSTM Forecasted Savings', color='orange')
plt.title('LSTM Forecast of Long-Term Savings Growth')
plt.xlabel('Date')
plt.ylabel('Savings Amount')
plt.legend()
plt.show()
```

Sentiment Analysis

Sentiment Analysis Overview
Sentiment analysis is a branch of natural language processing (NLP) that involves examining text data to determine the emotional tone behind it. In the context of employee feedback on the Two-Pot Retirement System, sentiment analysis can help understand how employees feel about the system, identify concerns, and gauge overall satisfaction. This information is critical for policymakers and financial institutions to make informed decisions about the retirement system's implementation and potential adjustments.

Steps in Sentiment Analysis
Data Collection:

The first step involves gathering employee feedback. This data could come from surveys, interviews, or open-ended responses in feedback forms regarding the Two-Pot Retirement System.
It is essential to ensure that the dataset is representative of various employee demographics, such as age, income level, and years of service, to capture diverse perspectives.
Data Preprocessing:

Cleaning the Text: Raw feedback often contains noise such as typos, punctuation, and irrelevant information. Preprocessing involves converting the text to lowercase, removing punctuation and special characters, and eliminating stopwords (common words like "and," "the," etc. that do not contribute significant meaning).
Tokenization: This process involves breaking down the feedback text into individual words or tokens. Tokenization helps in analyzing the frequency and sentiment associated with specific terms.
Sentiment Scoring:

After preprocessing, the next step is to score the sentiment of each piece of feedback. Several models and tools can perform sentiment analysis:
Lexicon-Based Approaches: Tools like VADER (Valence Aware Dictionary and sEntiment Reasoner) analyze text based on pre-defined sentiment lexicons. These lexicons assign sentiment scores to words, which are then aggregated to determine the overall sentiment of a text.
Machine Learning Models: Advanced sentiment analysis can employ machine learning techniques that train models on labeled datasets (where feedback is pre-categorized as positive, negative, or neutral). This approach requires more data and computational resources but can yield highly accurate results.
Categorization of Sentiment:

Sentiment scores can be categorized into three main groups:
Positive Sentiment: Feedback that expresses satisfaction, support, or enthusiasm regarding the Two-Pot System.
Negative Sentiment: Feedback that indicates dissatisfaction, concerns, or criticism of the system.
Neutral Sentiment: Feedback that does not express strong feelings either way and might be more informational or factual.
Data Visualization:

Visualizing sentiment data can provide intuitive insights. Common methods include:
Histograms: To show the distribution of sentiment scores.
Bar Charts: To illustrate the percentage of positive, negative, and neutral sentiments.
Word Clouds: To highlight frequently mentioned words in the feedback, which can indicate common themes or issues.
Insights and Reporting:

The results of the sentiment analysis can be summarized in a report that highlights key findings, such as:
The overall sentiment towards the Two-Pot Retirement System.
Common concerns or themes mentioned by employees (e.g., difficulty in understanding the system, financial security worries, etc.).
Recommendations for policymakers based on the feedback, such as improving communication about the system or adjusting regulations to address employee concerns.
Importance of Sentiment Analysis
Conducting sentiment analysis on employee feedback about the Two-Pot Retirement System serves several critical purposes:

Understanding Employee Attitudes: It provides insights into how employees perceive the system, allowing for a deeper understanding of their needs and concerns.
Identifying Areas for Improvement: By analyzing feedback, policymakers and financial institutions can identify specific issues that need addressing, enhancing the system's effectiveness and employee satisfaction.
Informed Decision-Making: Data-driven insights allow for better decision-making regarding the implementation and adjustment of the Two-Pot System, ultimately aiming to balance short-term needs with long-term financial security.
